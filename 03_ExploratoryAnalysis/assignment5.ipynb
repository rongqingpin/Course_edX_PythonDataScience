```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import andrews_curves as adrcur
```

```python
# load the data
fname = '/Users/pinqingkan/OneDrive/JobSearch/DataScience/PythonDataScience/DAT210x-master/Module3/Datasets/wheat.data'
X = pd.read_csv(fname, index_col = 0)
```

```python
X.columns
```

```python
Y = X.drop(labels = ['area', 'perimeter'], axis = 1)
Y.columns
```

```python
# Andrews curve
adrcur(Y, 'wheat_type', alpha = 0.7)
plt.show()
```

```python
# Z = Y.append(X[['area', 'perimeter']])
Z = pd.concat([Y, X[['area', 'perimeter']]], axis = 1)
Z.columns
```

```python
# Andrews curve
adrcur(Z, 'wheat_type', alpha = 0.7)
plt.show()
```
