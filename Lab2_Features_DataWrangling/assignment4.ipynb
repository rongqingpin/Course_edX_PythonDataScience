```python
import pandas as pd
import numpy as np
```

```python
# specify which rows to skip
rows1 = np.arange(12,100,12)
rows2 = np.arange(13,100,12)
rows = np.append(rows1, rows2)
rows = sorted(rows)
rows
```

```python
# load the data from the website
url = 'http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2'
X = pd.read_html(url, header = 1, skiprows = rows) #index_col = 0,
# X = pd.read_html(url)
X = X[0]

# drop the rows with > 4 NaN
# X = X.dropna(axis = 0, thresh = 4)
```

```python
# rename part of the columns
# X.rename(columns = {'G.1':'PP-G', "A.1":"PP-A", "G.2":"SH-G", "A.2":"SH-A"})
X.columns = ['RK','PLAYER','TEAM','GP','G','A','PTS','+/-','PIM','PTS/G','SOG','PCT','GWG','PP-G','PP-A','SH-G','SH-A']
```

```python
# get rid of 'RK' column
Y = X.drop(axis = 1, labels = ['RK'])
```
