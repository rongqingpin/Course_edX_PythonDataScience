#### scatter plot of a wheat dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data
flc = '/Users/pinqingkan/Desktop/Codes/Course_edX_PythonDataScience/03_ExploratoryAnalysis/Datasets/'
fname = flc + 'wheat.data'
X = pd.read_csv(fname, index_col = 0)
```

```python
X.columns
```

```python
# scatter plot
# plt.scatter(X.area, X.perimeter)
X.plot.scatter(x = 'area', y = 'perimeter', marker = 'o')
plt.show()
```

```python
X.plot.scatter(x = 'groove', y = 'asymmetry', marker = '^')
plt.show()
```

```python
X.plot.scatter(x = 'compactness', y = 'width', marker = '.')
plt.show()
```
